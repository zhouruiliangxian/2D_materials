```python
import matplotlib.pyplot as plt
import numpy as np
import itertools

def plot_confusion_matrix(cm, classes, title='Confusion Matrix', cmap_diag=None, cmap_other=None):
    plt.figure(figsize=(8, 8))
    confusion_matrix = np.array([[3104070,   40225,     990,    1675],
       [  54013, 3104070,   61260,   16225],
       [    510,   34898,  3104070,    4188],
       [   2445,   12454,   32065,  3104070]])
    # 使用自定义颜色映射
    plt.imshow(confusion_matrix, interpolation='nearest', cmap=cmap_other)
    plt.imshow(np.diag(np.diag(confusion_matrix)), interpolation='nearest', cmap=cmap_diag)
    
    tick_marks = np.arange(len(classes))
    
#     plt.title(title, fontsize=16)
#     plt.xlabel('Predicted', fontsize=14)
#     plt.ylabel('True', fontsize=14)
    
    plt.xticks([])
    plt.yticks([])
    
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        percentage = format(cm[i, j], '.2f') + '%'
        plt.text(j, i,percentage,
                 verticalalignment="center",
                 horizontalalignment="center",
                 color="white" if i==j else "black",
                 fontsize=20
                )

classes = ('Background', 'monolayer', 'bilayer','multilyayer')
plot_confusion_matrix(percentages, classes, cmap_diag=plt.cm.Blues, cmap_other=plt.cm.Reds)
plt.savefig('confusion.tif', dpi=300, bbox_inches='tight', pad_inches=0)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import itertools

def plot_confusion_matrix(cm, classes, title='Confusion Matrix', cmap=plt.cm.Blues):
    plt.figure(figsize=(10, 8))
    im = plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar(im, fraction=0.046, pad=0.04)  # Adds a colorbar with spacing adjustments
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    # Loop over data dimensions and create text annotations.
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, f'{cm[i, j]:.2f}',
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')

# Example data
cm = np.array([
    [3.1e+06, 51750, 213, 2736],
    [58858, 2.5e+06, 49142, 22346],
    [2196, 5e+05, 265149, 7117],
    [132, 12226, 33601, 119236]
])
classes = ['Class 0', 'Class 1', 'Class 2', 'Class 3']

plot_confusion_matrix(cm, classes)
plt.savefig('/mnt/data/confusion_matrix_with_colorbar.png', dpi=300)
```
