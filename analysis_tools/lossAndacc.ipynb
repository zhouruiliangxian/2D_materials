```python
df_train = pd.DataFrame()
df_loss = pd.DataFrame()
import pandas as pd
# 遍历 JSON 对象的列表
for each in json_list[:-1]:
    if 'aAcc' in each:
        try:
            # 确保 eval(each) 返回一个字典
            data = [eval(each)]
            df_train = pd.concat([df_train, pd.DataFrame(data)], ignore_index=True)
        except Exception as e:
            print(f"在追加到 df_test 时发生错误：{e}")
    else:
        try:
            # 确保 eval(each) 返回一个字典
            data = [eval(each)]
            df_loss = pd.concat([df_loss, pd.DataFrame(data)], ignore_index=True)
        except Exception as e:
            print(f"在追加到 df_train 时发生错误：{e}")
```
